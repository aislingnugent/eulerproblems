Fn = Fn-1 + Fn-2 where F1 = 1 F2 = 1

```python
def fib(n):
    
    fib_seq=[1,1]
    for i in range(2,n):
        fib_seq.append(fib_seq[i-2]+fib_seq[i-1])
    return fib_seq
```
