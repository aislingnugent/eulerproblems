```python
# find prime numbers below given number
import math as m
num = int(input('all prime up to:'))

factors=[]
for i in range(1, m.ceil(num/2)+2, 2):
    
    # check is i a factor
    if num % i == 0:
        
        
        factors.append(i)
```
