```python
sum_of_numbers = 0
for number in range(1,1000):
    if number%3 == 0 or number%5 == 0:
        sum_of_numbers += number
print(sum_of_numbers)
```

```python
fs = [0,1,1]
sum_fs_even = 0

while fs[len(fs)-1]<4000000:
    
    new_number = fs[len(fs)-1] + fs[len(fs)-2]
    
    fs.append(new_number)
    if new_number%2==0:
        sum_fs_even += new_number
        

print(sum_fs_even)
print(fs)
```
